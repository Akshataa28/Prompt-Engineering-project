```python
import pandas as pd
```

```python
prompts = [
"Prompt 1: Summarize the following text.",
"Prompt 2: Summarize the following text in 3 bullet points.",
"Prompt 3: Explain the following text in simple language.",
"Prompt 4: Extract key insights from the following text.",
"Prompt 5: Identify the main concepts from the following text.",
"Prompt 6: Rewrite the text in a concise way."
]

responses = [
"Data science uses statistics, programming, and domain knowledge to analyze data and produce insights.",

"- Data science uses statistics and programming.\n- It requires knowledge of a specific domain or field.\n- It helps analyze data and generate useful insights.",

"Data science is a field where people use math (statistics), computer programming, and knowledge of a specific subject to study data and find useful information from it.",

"Data science is an interdisciplinary field, It combines statistics, programming, and domain knowledge, Its main goal is to analyze data and generate insights.",

"Data Science, Statistics, Programming , Domain Knowledge, Data Analysis, Insights Generation",

"Data science uses statistics, programming, and domain knowledge to analyze data and gain insights."
]

data = []

for p, r in zip(prompts, responses):
    data.append({
        "Prompt": p,
        "Response": r
    })

df = pd.DataFrame(data)

df.to_excel("responses.xlsx", index=False)

print("Responses saved successfully.")
```
